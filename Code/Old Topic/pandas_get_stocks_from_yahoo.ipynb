# Get stock prices

```python
import pandas as pd
from pandas_datareader import data
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
print(f"pandas  version = {pd.__version__}")
print(f"seaborn version = {sns.__version__}")
```

### get All stocks

```python
allCompanyName = pd.read_csv("../Data/ListedCompanies/listedCompaniesNameOnly.csv")
allCompanyName.head()
```

```python
stocks=pd.Series(allCompanyName["หลักทรัพย์"]) + '.bk'
stocks = list(map(lambda e: e.lower(), stocks))
stocks
```

```python
df = data.DataReader(stocks[0:5], data_source="yahoo", #get >1 stock
                     start="2017-01-01")
df.head()
```

```python
df.info()
```

```python
df.index
```

```python
df['Close']
```

```python
sns.lineplot(x=df.index, y='2s.bk', data=df['Adj Close'])
```

```python
df['High']['2s.bk']
```

## stack data frame

```python
ds=df.stack()
ds
```

```python
ds = df.stack().reset_index()
ds = ds.rename_axis(None, axis=1)
ds = ds[['Date','Symbols','Close']]
# ds = ds.set_index('Date')
ds.tail(20)
```

```python
ds.isnull().sum()
```

```python
ds.to_csv('listedCompaniesDateNameClose.csv', index=False) 
```

```python
sns.lineplot(x='Date', y='Close', hue='Symbols',
             style='Symbols',
             data=ds);
```

```python
sns.relplot(x='Date', y='Adj Close', col='Symbols', col_wrap=3, #col='Symbols' = แยกรูปออกมาตาม col symbols, col_wrap=3 = print 3 graph ต่อ บรรทัด
            hue='Symbols',
            kind='line',
            height=5, # ความสูงของรูป
            facet_kws={'sharey': False}, #กราฟในแต่ละบรรทัด share แกน y กัน
            data=ds)
```

```python
sns.lineplot(x='Date', y='Adj Close', hue='Symbols',
             data=ds[ds.Symbols.str.match('7up|a')]) # ดูแค่ apple & meta
```
